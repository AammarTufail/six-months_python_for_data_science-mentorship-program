# **SVD (Singular value decomposition)**

A simple example of using SVD with sklearn in Python. The following example uses the UCI ML iris dataset (since the tips dataset isn't ideal for SVD). The sklearn.decomposition.TruncatedSVD class allows to perform linear dimensionality reduction.

```python
# import necessary libraries
from sklearn.decomposition import TruncatedSVD
from sklearn import datasets
import matplotlib.pyplot as plt

# load iris dataset as an example
iris = datasets.load_iris()
X = iris.data
y = iris.target

# SVD represent documents and terms in vectors 
svd = TruncatedSVD(n_components=2, algorithm='randomized', n_iter=100, random_state=122)
X_svd = svd.fit_transform(X)

# plot the result
plt.scatter(X_svd[:, 0], X_svd[:, 1], c=y, s=50, cmap='viridis')
plt.xlabel('Component 1')
plt.ylabel('Component 2')
plt.title('2-Component SVD')
plt.show()
```
