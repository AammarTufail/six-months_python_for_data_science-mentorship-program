## Kuch Popular APIs Jo Data Science aur ML Tasks Ke Liye Istemal Kiye Jaate Hain

1. **Twitter API**: Social media trends aur public opinion analyze karne ke liye.
2. **OpenWeatherMap API**: Weather related data hasil karne ke liye.
3. **Alpha Vantage API**: Stock market trends aur financial data analyze karne ke liye.
4. **Google Cloud Vision API**: Images ko analyze karke un mein maujood objects, text, aur emotions ko pehchanne ke liye.

## Kaise Shuru Karein?

APIs ke saath kaam karna shuru karne ke liye, aapko bas itna karna hai ke:

1. API ka documentation parhain.
2. API key ya authentication details hasil karein, agar zaroori ho.
3. Apne pasandida programming language (jaise Python) mein API se data request karein aur response ko handle karein.

## Nazar Dalte Hain Ek Example Par

Aaiye, ek simple Python script dekhte hain jo OpenWeatherMap API se Lahore ka current weather data fetch karta hai:

```python
import requests

API_KEY = "aapka_api_key_yahaan"
CITY = "Lahore"
URL = f"http://api.openweathermap.org/data/2.5/weather?q={CITY}&appid={API_KEY}"

response = requests.get(URL)
data = response.json()

print(f"Lahore ka temperature: {data['main']['temp']} Kelvin")
```

Yeh script Lahore ka current temperature fetch karke print karega, bilkul asaan!
